# Binary Search

## Question ($\star\star$, 30 marks)
- Problem: 
    - follow up on Exercise 3.15 in Chapter 3 of lecture note
    - find a target in a $m \times n$ matrix
    - here we can assume that 
        - each row in the matrix is sorted in ascending order
        - each column in the matrix is sorted in ascending order
        - $m < n$
    - see examples in the testing cases
    - find the solution with the complexity below
- Complexity: 
    - $O(m \times log(n)$ time
    - $O(1)$ space
- **Logic: # Implement me (5 marks)**


```python
# we use the binary search twice first over the rows (to find i), since items on row i are smaller than those on row i + 1
# Then over the columns (to find j), since items on each row are sorted in the matrix and return [i, j] 
# Implementation
def fun_21(matrix, target):
    """
    Find a target in a m * n matrix
    
    Parameters
    ----------
    matrix : list[list[int]] (i.e., a list of list of integers)
    target : an integer
    
    Returns
    ----------
    True  : if the target is in the matrix
    False : otherwise
    """
    
    # Implement me (25 marks)
    if len(matrix) == 0 or len(matrix[0]) == 0:
        return False
    col = len(matrix[0]) - 1
    for i in range(len(matrix)):
        while matrix[i][col] > target and col >= 0:
            col = col - 1
        if matrix[i][col] == target:
            return True
    return False
```

## Question ($\star\star$, 20 marks)
- Problem: 
    - follow up on Exercise 5.5 in Chapter 5 of lecture note
    - find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    - here, the path starts from the root and ends at a leaf node
    - see examples in the testing cases
    - find the solution with the complexity below    
    - you must use DFS to design the solution
- Complexity: 
    - $O(n)$ time
    - $O(n)$ space
    - where $n$ is the number of nodes in the tree
- **Logic: # Implement me (5 marks)**


```python
# we use a global variable, paths, to keep track of the paths
# in the helper function, we use a string, path, to keep track of each path
# when the root is None, return
# when the root is a leaf
# add the value of the root to path
# append path to paths
# return
# when the root is not a leaf
# add the value of the root and the arrow ('->') to path
# recursively apply the helper function to the left and right subtree
# Implementation
def fun_31(root, target):
    """
    Find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    
    Parameters
    ----------
    root : the root of a binary tree
    target : an integer
    
    Returns
    ----------
    True,  if there is such a path
    False, otherwise
    """
    
    # Implement me (15 marks)
    if root is None:
        return False

    return True if helper(root, target) else False

class TreeNode(object):

    def __init__(self, x):
        # The value of the node
        self.val = x
        # The left child
        self.left = None
        # The right child
        self.right = None

def helper(root, target):
    if root is None:
        return False
    if root.val == target and root.left is None and root.right is None:
        return True

    return True if helper(root.left, target - root.val) or helper(root.right, target - root.val) else False


```

```python
# Reference: Part of the code below is from chapter_5_tree_search_solution.ipynb
# we use a variable, count, to keep track of the number of 1s
# we loop over the matrix and for each 1 we:
# increase count by 1
# use DFS to remove the connected 1s

# Implementation
def fun_32(matrix):
    """
    Find the number of (vertically or horizontally) connected deserts in a map (represented by a matrix)
    
    Parameters
    ----------
    matrix : a matrix where 1 denotes desert and 0 denotes grassland
    
    Returns
    ----------
    the number of connected deserts
    """

    # Corner case
    if len(matrix) == 0 or len(matrix[0]) == 0:
        return 0

    m, n = len(matrix), len(matrix[0])
    # The number of connected deserts
    count = 0

    for i in range(m):
        for j in range(n):
            if matrix[i][j] == 1:
                count += 1
                # DFS: remove '1's conntected to entry [i, j]
                helper(matrix, m, n, i, j)
                
    # Implement me (5 marks)
    # Restore the matrix

    
    return count
```

```python
def helper(matrix, m, n, i, j):
    """
    DFS: remove '1's conntected to entry [i, j]

    Parameters
    ----------
    matrix : a matrix where 1 denotes desert and 0 denotes grass land
    m : the nubmer of rows in the matrix
    n : the nubmer of columns in the matrix
    i, j : entry [i, j]
    """

    # Implement me (5 marks)
    # If either i or j is not valid, or matrix[i][j] is not 1
    if i < 0 or i >= m or j < 0 or j >= n or matrix[i][j] != 1:
        return

    # Remove 1
    matrix[i][j] = -1

    # Remove '1' above [i, j]
    helper(matrix, m, n, i - 1, j)
    # Remove '1' below [i, j]
    helper(matrix, m, n, i + 1, j)
    # Remove '1' on the left-hand side of [i, j]
    helper(matrix, m, n, i, j - 1)
    # Remove '1' on the right-hand side of [i, j]
    helper(matrix, m, n, i, j + 1)
```

# Breadth First Search (35 marks)

## Question ($\star\star$, 20 marks)
- Problem: 
    - follow up on Question 3.1 above
    - find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    - see examples in the testing cases
    - find the solution with the complexity below
    - you must use BFS to design the solution
    - the node value can be modified <font color='red'>(you may consider this as a hint)</font>
- Complexity: 
    - $O(n)$ time
    - $O(n)$ space
- **Logic: # Implement me (5 marks)**


```python
# if the root is None (empty), return
# otherwise we:
# display the value of the root
# traverse the left subtree by recursively calling the pre-order function
# traverse the right subtree by recursively calling the pre-order function

# Implementation
from collections import deque

class TreeNode(object):

    def __init__(self, x):
        # The value of the node
        self.val = x
        # The left child
        self.left = None
        # The right child
        self.right = None

def fun_41(root, target):
    """
    Find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    
    Parameters
    ----------
    root : the root of a binary tree
    target : an integer
    
    Returns
    ----------
    True,  if there is such a path
    False, otherwise
    """
    
    # Implement me (15 marks)
    if root is None:
        return False
    # Initialize the queue
    queue = deque([root.left, root.right])
    values = deque([root.val])
    values.append(root.val)
    # BFS
    while len(queue) > 0:
        node_left, node_right = queue.popleft(), queue.popleft()

        if node_left == None and node_right == None:
            continue

        if node_left is not None:
            node_left.val = node_left.val + values.popleft()
            queue.append(node_left.left)
        if node_right is not None:
            node_right.val = node_right.val + values.popleft()
            queue.append(node_right.right)
        if node_left is not None:
            values.append(node_left.val)
            queue.append(node_left.right)
        if node_right is not None:
            queue.append(node_right.left)
            values.append(node_right.val)

    while len(values) > 0:
        if values.popleft() == target:
            return True
    return False


```

## Question ($\star$) (15 marks)
- Problem:
    - follow up on Exercise 6.10 in Chapter 5 of lecture note
    - unlike Exercise 6.10 where we should find the bottom-left node in a binary tree, here we should find the bottom-right node
    - see examples in the test cases
    - find the solution with the complexity below
    - you must use BFS to design the solution
- Complexity:
    - $O(n)$ time
    - $O(n)$ space
- **Logic: # Implement me (5 marks)**


```python
# we use a variable, left_most, to keep track of the value of the left-most node on each level
# after traversing all the nodes, left_most will be the value of the bottom-left node

# Implementation

class TreeNode(object):

    def __init__(self, x):
        # The value of the node
        self.val = x
        # The left child
        self.left = None
        # The right child
        self.right = None
        
def fun_42(root):
    """
    Find the bottom-right node in a binary tree
    
    Parameters
    ----------
    root : the root of a tree
    
    Returns
    ----------
    The value of the bottom-right node in a binary tree
    """

    # Implement me (10 marks)
    if root is None:
        return root
    bottom_right = root.val

    queue = deque([root])

    while len(queue) > 0:
        n = len(queue)
        for i in range(n):
            node = queue.popleft()

            # Update bottom_left using the value of the left-most node on each level
            if i == 0:
                bottom_right = node.val

            if node.right is not None:
                # Add the value of the right child to the end of the queue
                queue.append(node.right)
            if node.left is not None:
                # Add the value of the left child to the end of the queue
                queue.append(node.left)

    return bottom_right

 
```
